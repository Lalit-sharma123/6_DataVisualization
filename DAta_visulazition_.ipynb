Question 5: Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap

```python
# Set the random seed for reproducibility
np.random.seed(42)

# Generate correlated data
size = 100  # Number of samples
x = np.random.normal(0, 1, size)
y = 2 * x + np.random.normal(0, 1, size)
z = 0.5 * x - y + np.random.normal(0, 1, size)

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y, 'Z': z})

correlation_matrix = data.corr()


sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Matrix Heatmap")
plt.show()

```

Question 1:Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')
p.line(x, y, legend_label="sin(x)", line_width=2)

show(p)
```

Question 2: 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
import numpy as np

output_notebook()

np.random.seed(42)
x = np.random.rand(100) * 100
y = np.random.rand(100) * 100
sizes = np.random.rand(100) * 50
colors = np.random.randint(1, 256, size=(100, 3))

data = {
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': ['#%02x%02x%02x' % (r, g, b) for r, g, b in colors]
}

source = ColumnDataSource(data=data)

p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y')
p.scatter('x', 'y', size='sizes', color='colors', fill_alpha=0.6, source=source)

show(p)

```

Question 3:3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset

```python
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

data = {'fruits': fruits, 'counts': counts}
source = ColumnDataSource(data=data)
p = figure(x_range=fruits, title="Fruit Counts",
           toolbar_location=None, tools="")

p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', fill_color='blue')

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Counts"
p.xaxis.axis_label = "Fruits"
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

show(p)
```

Question 5:  Create a Bokeh heatmap using the provided dataset

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import transform
import numpy as np

output_notebook()

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
}

p = figure(title="Heatmap Example", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1), tools="", toolbar_location=None)

mapper = LinearColorMapper(palette="Viridis256", low=data['value'].min(), high=data['value'].max())

p.rect(x='x', y='y', width=0.1, height=0.1, source=data,
       fill_color=transform('value', mapper), line_color=None)

color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)

```
